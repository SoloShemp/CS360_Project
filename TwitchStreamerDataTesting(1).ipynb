```python
import pandas as pd
import time as time
import numpy as np
import matplotlib.pyplot as plt
import tkinter as tk
from tkinter import filedialog
from sklearn.preprocessing import MinMaxScaler


# You can borrow this for the rest of the semester to not have to worry about paths to your data
# you will see an example use case in problem 4
def open_file_dialog():
    root = tk.Tk()
    root.withdraw()
    return tk.filedialog.askopenfilename()

%matplotlib inline
```

```python
df = pd.read_csv(open_file_dialog())
```
